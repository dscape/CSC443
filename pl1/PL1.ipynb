```python
import subprocess
from datetime import datetime
```

```python
KB = 1024
MB = 1024 * KB

file_size = 50.0 * MB
blocks = [10, 100, KB, 5*KB, 11*KB, 101*KB, 301*KB, 501*KB, MB, 2*MB, 3*MB]
timings_dd = []
timings_cpp = []
```

```python
# Measure timings of sequential write using dd and create_random_file
for block_size in blocks:
    count = int(ceil(file_size/block_size))
    
    start = datetime.now()
    subprocess.call(['dd', 'if=/dev/urandom', 'of=garbage-dd-{0}'.format(block_size), 'bs={0}'.format(block_size), 'count={0}'.format(count)])
    end = datetime.now()
    timings_dd.append((end-start).total_seconds() * 1000)
    
    start = datetime.now()
    subprocess.call(['./create_random_file', 'garbage-{0}'.format(block_size), '{0}'.format(file_size), '{0}'.format(block_size)])
    end = datetime.now()
    timings_cpp.append((end-start).total_seconds() * 1000)
```

```python
semilogx(blocks, timings_dd, 'bo-', label='dd')
semilogx(blocks, timings_cpp, 'ro-', label='c++')
ylabel('milliseconds')
xlabel('block size')
title('File: 50MB, Disk: Solid State')
legend()
```
