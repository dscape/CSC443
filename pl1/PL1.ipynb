```python
KB = 1024
MB = 1024 * KB

file_size = 50.0 * MB
blocks = [10, 100, KB, 5*KB, 11*KB, 101*KB, 301*KB, 501*KB, MB, 2*MB, 3*MB]
timings = []
```

```python
# Measure timings of sequential write using dd
for block_size in blocks:
    count = int(ceil(file_size/block_size))
    start = datetime.now()
    subprocess.call(['dd', 'if=/dev/urandom', 'of=garbage-{0}'.format(block_size), 'bs={0}'.format(block_size), 'count={0}'.format(count)])
    end = datetime.now()
    timings.append((end-start).total_seconds() * 1000)
```

```python
loglog(blocks, timings, 'bo-', label='dd')
ylabel('milliseconds')
xlabel('block size')
title('File: 50MB, Disk: Solid State')
legend()
```
